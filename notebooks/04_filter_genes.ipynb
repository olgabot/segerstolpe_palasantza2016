```python
import pandas as pd
```

```python
rpkm = pd.read_table("/home/ecwheele/cshl2017/segerstolpe_palasantza2016/data/01_subset_metadata/pancreas_rpkm_792cells_clean.txt",
                    index_col=0)
rpkm.head()
```

```python
expressed_genes = pd.read_excel("/home/ecwheele/cshl2017/segerstolpe_palasantza2016/data/00_original/mmc2.xlsx",
                               sheetname = 'ExpressedGenes_Celltypes', skiprows = 4, skipcols=2)
expressed_genes.head()
```

```python
#Only going to keep top expressed genes in the cells we are interested in. 
#First make a list of cells we are interested in (these are our column names)

cells_to_keep = ['Î±-cells','Î²-cells','Î³-cells','Î´-cells','Îµ-cells','unclass endocrine']

#Subset the expressed genes matrix for only those columns with a square bracket after
#the dataframe name
expressed_genes_endocrine = expressed_genes[cells_to_keep]
print(expressed_genes_endocrine.shape)
expressed_genes_endocrine.head()
```

```python
#Make a list of genes to keep. We want to keep the top 500 genes in each dataframe. 

genes_to_keep = []

#We are going to loop through each of the columns with a for loop
for i in expressed_genes_endocrine.columns:
    #Slice that column in the dataframe to grab the first 500 rows
    genes = expressed_genes_endocrine[i][:500]
    #Get rid of the extra quotation around the gene names 
    genes = [x.strip("'") for x in genes]
    #Extend our master list with the new genes that we have identified.
    genes_to_keep.extend(genes)
    
print(len(genes_to_keep))

#Use a set object to get a unique list. Set will drop duplicates in the list.
unique_genes = set(genes_to_keep)
print(len(unique_genes))
```

```python
#Now only keep genes in the counts matrix that were ones that we identified above. 
#Since gene ID is stored as the row name, we are going to access those rows with .loc
#followed by the list of genes names we want

rpkm_filtered = rpkm.loc[unique_genes]
print(rpkm_filtered.shape)
rpkm_filtered.head()
```

```python
#Since we are using gene names, lets drop that gene ID column. 

#use axis=1 to say we are dropping from the columns and inplace=True to replace 
#rpkm_filtered with the new 

rpkm_filtered.drop('1', axis=1, inplace=True)
rpkm_filtered.head()
```

```python
rpkm_filtered.to_csv("/home/ecwheele/cshl2017/segerstolpe_palasantza2016/data/01_subset_metadata/rpkm_filter_cells_and_genes.txt",
                    sep="\t", encoding = 'utf-8')
```
